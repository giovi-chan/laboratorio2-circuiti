```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Aprire il file in modalità di lettura
with open('err_casuale ampiezza 1k.txt', 'r') as file:
    # Leggere tutte le righe del file
    lines = file.readlines()

# Inizializzare le liste per memorizzare i dati
x_values = []
y_values = []

# Iterare attraverso le righe e estrarre i dati
for line in lines:
    # Dividere la riga in base agli spazi
    parts = line.split()
    # Convertire i valori in numeri reali e aggiungerli alle rispettive liste
    x_values.append(float(parts[0]))
    y_values.append(float(parts[1]))

# Definire la funzione gaussiana
def gaussian(x, amplitude, mean, sigma):
    return amplitude * np.exp(-((x - mean) / sigma) ** 2 / 2)

# Calcolare l'istogramma dei dati
num_bins = 40

hist, bin_edges = np.histogram(y_values, bins=num_bins, density=True)
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])

# Esegui il fit della gaussiana
param_bounds = ([0, min(y_values), 0], [3500, max(y_values), 1])
popt, pcov = curve_fit(gaussian, bin_centers, hist, bounds=param_bounds, maxfev =70000)

# Estrai i parametri ottimizzati
amplitude_fit, mean_fit, stddev_fit = popt

# Visualizza i dati e il fit
plt.hist(y_values, bins=num_bins, density=True, label='Data')
plt.plot(bin_centers, gaussian(bin_centers, *popt), color='red', label='Gaussian Fit')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Fit con una funzione gaussiana')
plt.legend()
plt.show()

# Calcola i valori predetti dal modello ai punti x_values
y_predicted = gaussian(x_values, *popt)

# Calcola l'errore associato a ciascun punto (puoi usare la deviazione standard dei dati)
errors = stddev_fit
print(stddev_fit)
print(mean_fit)

# Calcola il chi-quadro
chi_squared = np.sum(((y_values - y_predicted) / errors)**2)

print("Chi-squared:", chi_squared)

```
