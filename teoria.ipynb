```python
import numpy as np
import matplotlib.pyplot as plt

R = (329.84, 470.53, 997.90, 9970.9)
C = 326.1E-9
L = 472.5E-6

for i in range(4):  # Changed the range to iterate over all elements of R
    H_values = []  # Empty list to store H values for each frequency
    H2_values = []
    for f in range(8000, 18001):  # Changed the range to start from 1
        A = 2 * f * np.pi * L / (1 - 2 * f * np.pi * 2 * f * np.pi * L * C)
        H = (R[i] / np.sqrt(R[i] * R[i] + A * A))*0.1
        H2= R[i]*R[i]/(R[i] * R[i] + A * A)
        H_values.append(H)    # Append H to the list
        H2_values.append(H2)   
        
    
        
    plt.figure(1)
    plt.plot(range(8000, 18001), H_values, label=f'R = {R[i]}')  # Plot H values for current R
    plt.xlabel('Frequency')
    plt.ylabel('H')
    plt.title('Frequency Response-module')
    plt.legend()
    
    plt.figure(2)
    plt.plot(range(8000, 18001), H2_values, label=f'R = {R[i]}')  # Plot H values for current R
    plt.xlabel('Frequency')
    plt.ylabel('H2')
    plt.title('Frequency Response-real part')
    plt.legend()
plt.show()
```
